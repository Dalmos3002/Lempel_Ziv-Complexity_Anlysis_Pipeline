# Data betöltés .mat-ban konvertálás .npy-ba

```python
# Használt könyvtárak
import os
import numpy as np
import pandas as pd
import h5py as hp
```

```python
# Data.mat betöltése
MAT_data = hp.File(r'D:\Pt02_Full_data\Pt02.DryadRealignedLFP.mat')

variable_names = list(MAT_data.keys())
print("Változók a fájlban:", variable_names)
```

```python
# Azonnali dataset keresés
for name in MAT_data:
    obj = MAT_data[name]
    if isinstance(obj, hp.Dataset):
        print(f"Dataset található: '{name}' - {obj.shape}")
        NP_data = MAT_data[name][:]  				# Adatok kinyerése
        print("Conversion to .npy is done!")
        break
else:
    print("Nem található dataset")
```

```python
# Ha sikerült betölteni
if 'NP_data' in locals():
    print(f"Adat információk:")
    print(f"Alak: {NP_data.shape}")
    print(f"Típus: {NP_data.dtype}")
    cleanDataFS = MAT_data['cleanDataFS'][()]  
    print("cleanDataFS értéke:", cleanDataFS)
    print(f"Percekben: {NP_data.shape[0] / 1250 :.2f} másodperc")
    print(f"Percekben: {NP_data.shape[0] / 1250 / 60:.2f} perc")
    print(f"Adat memóriafoglalás: {NP_data.nbytes / (1024**3):.2f} GB")
```

```python
# MAT bezárása
MAT_data.close()
print("Done!")
```

```python
# Mentés
np.save(r'D:\LZC_Analysis\Pt02_LZC\Pt02_LFP.npy', NP_data)
```
