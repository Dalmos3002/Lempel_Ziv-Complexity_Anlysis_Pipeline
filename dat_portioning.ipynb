# Data elharmadolása

```python
# Használt könyvtárak
import numpy as np
```

```python
# Paraméterek megadása
sampling_rate = 1250
total_seconds = 645
total_samples = sampling_rate * total_seconds

# Pontosan 3 egyenlő rész
samples_per_third = total_samples // 3
```

```python
# Feldarabolandó data betöltése
LFP_20nth = np.load(r'D:/LZC_Analysis/Pt02_LZC/Pt02_LFP_every_20th.npy')
print("Alap adat alakja (shape):", LFP_20nth.shape)
```

```python
# Darabolás
NP_data_20nth_1st_third = LFP_20nth[:samples_per_third, :]
NP_data_20nth_2nd_third = LFP_20nth[samples_per_third:samples_per_third*2, :]
NP_data_20nth_3rd_third = LFP_20nth[samples_per_third*2:, :]

print("Első adatharmad alakja (shape):", NP_data_20nth_1st_third.shape)
print("Második adatharmad alakja (shape):", NP_data_20nth_2nd_third.shape)
print("Harmadik adatharmad adat alakja (shape):", NP_data_20nth_3rd_third.shape)
```

```python
# Új adatok mentése
np.save(r'D:\LZC_Analysis\Pt02_LZC\Pt02_LFP_20nth_1st_third.npy', NP_data_20nth_1st_third)
np.save(r'D:\LZC_Analysis\Pt02_LZC\Pt02_LFP_20nth_2nd_third.npy', NP_data_20nth_2nd_third)
np.save(r'D:\LZC_Analysis\Pt02_LZC\Pt02_LFP_20nth_3rd_third.npy', NP_data_20nth_3rd_third)
print("Done!")
```
