```python
# Minden 20. csatorna kiválasztása (n=20)

def select_every_nth_channel(data, n=20):
    #Minden n-edik csatorna kiválasztása
    total_channels = data.shape[1]
    selected_indices = list(range(0, total_channels, n))
    
    # Csak a kiválasztott csatornák kinyerése
    selected_data = data[:, selected_indices]
    
    print(f"Csatorna szelektálás:")
    print(f"   • Eredeti csatornák: {total_channels}")
    print(f"   • Kiválasztott csatornák: {len(selected_indices)}")
    print(f"   • Csatorna indexek: {selected_indices}")
    print(f"   • Új adat alak: {selected_data.shape}")
    
    return selected_data, selected_indices

# Csatornák kiválasztása
reduced_data, channel_indices = select_every_nth_channel(full_data, n=20)
```
