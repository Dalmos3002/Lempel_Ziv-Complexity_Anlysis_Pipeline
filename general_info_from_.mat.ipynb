# Data betöltés .mat-ban + general infó kinyerés

```python
# Használt könyvtárak
import os
import numpy as np
import pandas as pd
import h5py as hp
```

```python
# Data.mat betöltése
MAT_data = hp.File(r'D:\Pt01_Full_data\Pt01.DryadRealignedLFP.mat')

variable_names = list(MAT_data.keys())
print("Változók a fájlban:", variable_names)
```

```python
# Változók kiíratása
print("\n=== VÁLTOZÓK RÉSZLETEI ===")

cleanData_dataset = MAT_data['cleanData']
print(f"cleanData: {cleanData_dataset.shape} - {cleanData_dataset.dtype}")

var_names = ['cleanDataFS', 'noiseFreqs', 'pulseFreqs', 'stable_time']
for name in var_names:
    obj = MAT_data[name]
    if isinstance(obj, hp.Dataset):
        print(f"{name}: {obj.shape} - {obj.dtype} - érték: {obj[:]}")
    else:
        print(f"{name}: {type(obj)} (nem Dataset)")


# Fájl bezárása
# MAT_data.close()
```
